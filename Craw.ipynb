```python
import requests
from bs4 import BeautifulSoup
```

```python
def extract_list(url,post_id):
    url_list = []
    bad_url_list = []
    get_post = False
    page_idx = 1
    #post_id = 11000
    while not get_post:
        #page_url = ("http://www.airitilibrary.com/Search/alThesisbrowse?"
        #"FirstID=U0001&ThirdID=D0001009002&type=Dissertations&SecondID=C0001009"
        #"&publicationTypeID=publicationType_all&searchConditionJson=%255B%255D"
        #"&page={}".format(page_idx))
        
        page_url = (url+str(page_idx))
        #&searchYearJson=%255B%25222016%2522%252C%25222018%2522%255D
        page_req = requests.get(page_url).text
        
        #print(page_url)
        #print(page_req)
        
        page_soup = BeautifulSoup(page_req, 'html.parser')
        if len(page_soup.select('span.title_con1 a')) == 0:
            print ("end")
            get_post = True
        for a in page_soup.select('span.title_con1 a'):
            if '/Publication/Index/' in a.get('href'):
                post_url = "http://www.airitilibrary.com" + a.get('href').strip()
                #print(post_url)
                if post_url not in url_list and post_url not in bad_url_list:
                    valid_post = extract_content(post_url, post_id)
                    if valid_post:
                        url_list.append(post_url)
                        post_id += 1
                    else:
                        bad_url_list.append(post_url)
                    if len(url_list) == 2281:
                    #if len(bad_url_list)==20:     
                        get_post = True
                        break
        if not get_post:
            page_idx += 1
        print (len(url_list), len(bad_url_list), page_idx)
    print('Finish',len(url_list),len(bad_url_list))
```

```python
# list_url = "http://www.airitilibrary.com/Search/alThesisbrowse?TabID=alThesisbrowse&PublicationTypeID=publicationType_all&FirstID=U0001&SecondID=C0001009&ThirdID=D0001009002&pOrderby=&type=Dissertations&page=1&title=&changeColor=&searchConditionJson=%255B%255D&searchAreaJson=&searchDegreeJson=&searchYearJson=%255B%25222016%2522%252C%25222018%2522%255D&page=1"
# txt = requests.get(list_url).text
# soup = BeautifulSoup(txt, 'html.parser')
# for a in soup.select('span.title_con1 a'):
#     post_url = "http://www.airitilibrary.com" + a.get('href')
#     print (post_url)

def extract_content(post_url, post_id):
    output_id = str(post_id).zfill(6)
    #print(output_id)
    post_text = requests.get(post_url).text
    post_soup = BeautifulSoup(post_text, 'html.parser')
    zh_title = post_soup.find('meta', property='og:title').get('content').strip()
    zh_content = ""
    if post_soup.select('textarea.TextArea1'):
        zh_content = post_soup.select('textarea.TextArea1')[0].text
    en_title = ""
    for titles in post_soup.select('div.booksInformation p a'):
        if titles.get('style') and 'text-decoration' in titles.get('style'):
            if titles.text.strip() and titles.text.strip() != zh_title:
                en_title = titles.text.strip()
                break
    en_content = ""
    if post_soup.select('textarea.TextArea2'):
        en_content = post_soup.select('textarea.TextArea2')[0].text

    if zh_title and zh_content and en_title and en_content:
        zh_title = zh_title.replace('\n', '')
        en_title = en_title.replace('\n', '')
        #print(zh_content)
        #print(en_content)
        try:
            with open("./Abstract_Pubhealth/{}c.txt".format(output_id), "w") as output:
                output.write('%s\n' % zh_title)
                output.write('%s' % zh_content)
            output.close()
            with open("./Abstract_Pubhealth/{}e.txt".format(output_id), "w") as output:
                output.write('%s\n' % en_title)
                output.write('%s' % en_content)
            output.close()
            print (post_id, zh_title, en_title)
            return True
        except Exception:
            print ('yabai')
            return False
    else:
        return False
```
